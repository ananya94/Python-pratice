```python
def datachange(row):
    ans  = ('terrible' in row['Text'].lower())
    return ans
```

```python
mydataframe['bad review'] = mydataframe.apply(datachange, axis = 1)
mydataframe
```

```python
mydataframe['bad review'].value_counts()
```

```python
bad_review_subset = mydataframe[mydataframe['bad review'] == True]
```

```python
bad_review_subset = bad_review_subset.groupby('Score' , as_index=False).count()
```
